```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from bs4 import BeautifulSoup
import pandas as pd
import time
import random

# ---------------------------
# Make Undetectable Driver
# ---------------------------
def get_driver():
    options = Options()
    options.add_argument("--disable-blink-features=AutomationControlled")
    options.add_argument("user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/114 Safari/537.36")

    driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)

    # anti-detect
    driver.execute_cdp_cmd(
        "Page.addScriptToEvaluateOnNewDocument",
        {
            "source": """
                Object.defineProperty(navigator, 'webdriver', {get: () => undefined});
            """
        }
    )
    return driver

driver = get_driver()

# ---------------------------
# Category link
# ---------------------------
url = "https://www.banggood.com/Wholesale-Electronics-ca-2001.html"

all_data = []

print("Opening page...")
driver.get(url)
time.sleep(random.uniform(3,5))

# scroll to load products
for _ in range(10):
    driver.execute_script("window.scrollBy(0, document.body.scrollHeight);")
    time.sleep(1.5)

# parse page
soup = BeautifulSoup(driver.page_source, "html.parser")

# correct selector
products = soup.select(".good-item")
if not products:
    products = soup.select("li[data-sku]")

print("Found:", len(products), "products")

# extract data
for p in products:
    try:
        name = p.select_one(".title").get_text(strip=True)
        price = p.select_one(".price").get_text(strip=True)
        link = p.select_one("a")["href"]

        if not link.startswith("http"):
            link = "https://www.banggood.com" + link

        all_data.append({
            "Name": name,
            "Price": price,
            "URL": link
        })
    except:
        pass

driver.quit()

# ---------------------------
# Save CSV
# ---------------------------
df = pd.DataFrame(all_data)
df.to_csv("banggood_data.csv", index=False, encoding="utf-8-sig")

print("\nDone! Saved:", len(df), "items → banggood_data.csv")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import io

# ==========================================
# PART 0: MOCK DATA GENERATION (FOR TESTING)
# ==========================================
# If you have your real 'banggood_scraped_data.csv', 
# set use_real_file = True below.
use_real_file = False

if not use_real_file:
    print("⚠ Using MOCK DATA for demonstration...")
    csv_data = """Category,Name,Price,Rating,Reviews,URL
Electronics,Wireless Earbuds,US$25.99,4.8,1.2k reviews,http://example.com/1
Electronics,Generic Cable,US$2.50,3.5,45 reviews,http://example.com/2
Smartphones,Redmi Note 12,US$199.00,4.5,500 reviews,http://example.com/3
Smartphones,iPhone 13 Case,US$5.99,N/A,N/A,http://example.com/4
Home & Garden,Smart LED Bulb,US$12.50,4.9,2.5k reviews,http://example.com/5
Home & Garden,Garden Hose,US$35.00,4.2,120 reviews,http://example.com/6
Lights,Solar Garden Light,US$15.99,4.6,89 reviews,http://example.com/7
Lights,Desk Lamp,US$45.00,4.7,1.1k reviews,http://example.com/8
Projectors,Mini Projector 4K,US$150.00,4.3,300 reviews,http://example.com/9
Electronics,Bluetooth Speaker,US$45.50,4.4,210 reviews,http://example.com/10
"""
    df = pd.read_csv(io.StringIO(csv_data))
else:
    # Load your actual scraped file
    try:
        df = pd.read_csv("banggood_scraped_data.csv")
        print("✅ Real data loaded successfully.")
    except FileNotFoundError:
        print("❌ File not found. Please ensure 'banggood_scraped_data.csv' is in the folder.")
        exit()

# ==========================================
# PART 2: DATA CLEANING & TRANSFORMATION
# ==========================================

print("\n--- Raw Data Head ---")
print(df.head())

# 1. Clean Price
# Removes 'US$', 'PKR', commas, and converts to float
def clean_price(price_str):
    if pd.isna(price_str) or price_str == 'N/A':
        return np.nan
    # Remove all non-numeric characters except the dot
    clean_str = str(price_str).replace('US$', '').replace(',', '').strip()
    try:
        return float(clean_str)
    except ValueError:
        return np.nan

# 2. Clean Reviews
# Handles '1.2k', '500 reviews', 'N/A'
def clean_reviews(review_str):
    if pd.isna(review_str) or review_str == 'N/A':
        return 0
    
    review_str = str(review_str).lower().replace('reviews', '').replace('review', '').strip()
    
    if 'k' in review_str:
        try:
            # Convert 1.2k -> 1200
            number = float(review_str.replace('k', ''))
            return int(number * 1000)
        except ValueError:
            return 0
    
    try:
        return int(review_str)
    except ValueError:
        return 0

# 3. Clean Rating
def clean_rating(rating_str):
    if pd.isna(rating_str) or rating_str == 'N/A':
        return np.nan # Keep as NaN to fill later or drop
    try:
        return float(str(rating_str).replace(' stars', '').strip())
    except ValueError:
        return np.nan

# Apply Cleaning
df['Price_Clean'] = df['Price'].apply(clean_price)
df['Reviews_Clean'] = df['Reviews'].apply(clean_reviews)
df['Rating_Clean'] = df['Rating'].apply(clean_rating)

# Handle Missing Values
# Fill missing ratings with the Category average (Imputation)
df['Rating_Clean'] = df['Rating_Clean'].fillna(df.groupby('Category')['Rating_Clean'].transform('mean'))
# If still NaN (e.g., category has no ratings), fill with global mean
df['Rating_Clean'] = df['Rating_Clean'].fillna(df['Rating_Clean'].mean())

# Drop rows where Price is missing (critical data)
df.dropna(subset=['Price_Clean'], inplace=True)

# ------------------------------------------
# Transformation: Create Derived Features
# ------------------------------------------

# Feature 1: Revenue Potential (Estimation)
# Assumption: A conservative 10% of buyers leave a review
df['Est_Revenue'] = df['Price_Clean'] * (df['Reviews_Clean'] * 10)

# Feature 2: Price Tier
# Categorize items into Budget, Mid-Range, Premium based on quantiles
q33 = df['Price_Clean'].quantile(0.33)
q66 = df['Price_Clean'].quantile(0.66)

def categorize_price(price):
    if price < q33: return 'Budget'
    elif price < q66: return 'Mid-Range'
    else: return 'Premium'

df['Price_Tier'] = df['Price_Clean'].apply(categorize_price)

print("\n--- Cleaned Data Head ---")
print(df[['Category', 'Name', 'Price_Clean', 'Reviews_Clean', 'Rating_Clean', 'Price_Tier']].head())

# ==========================================
# PART 3: EXPLORATORY ANALYSIS (5 Analyses)
# ==========================================
sns.set_theme(style="whitegrid")

# Analysis 1: Price Distribution per Category (Boxplot)
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Category', y='Price_Clean', palette='Set2')
plt.title('1. Price Distribution by Category')
plt.ylabel('Price (USD)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Analysis 2: Rating vs. Price Correlation (Scatter)
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Price_Clean', y='Rating_Clean', hue='Category', s=100, alpha=0.7)
plt.title('2. Do expensive items have higher ratings?')
plt.xlabel('Price (USD)')
plt.ylabel('Rating (0-5)')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()

# Analysis 3: Top 5 Most Reviewed Products
top_reviewed = df.nlargest(5, 'Reviews_Clean')
plt.figure(figsize=(10, 6))
sns.barplot(data=top_reviewed, y='Name', x='Reviews_Clean', hue='Category', dodge=False)
plt.title('3. Top 5 Most Reviewed Products')
plt.xlabel('Number of Reviews')
plt.ylabel('')
plt.tight_layout()
plt.show()

# Analysis 4: Best Value Metric
# Formula: (Rating / Price) * 100 -> Higher is better value
# High rating + Low price = High Score
df['Value_Score'] = (df['Rating_Clean'] / df['Price_Clean']) * 10
best_value = df.loc[df.groupby('Category')['Value_Score'].idxmax()]

print("\n--- 4. Best Value Item Per Category ---")
print(best_value[['Category', 'Name', 'Price_Clean', 'Rating_Clean', 'Value_Score']])

# Analysis 5: Average Revenue Potential by Category
avg_rev = df.groupby('Category')['Est_Revenue'].mean().sort_values(ascending=False).reset_index()

plt.figure(figsize=(10, 6))
sns.barplot(data=avg_rev, x='Category', y='Est_Revenue', palette='viridis')
plt.title('5. Average Estimated Revenue per Product by Category')
plt.ylabel('Est. Revenue ($)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# ==========================================
# FINAL: Save Cleaned Data
# ==========================================
df.to_csv("banggood_cleaned_data.csv", index=False)
print("\n✅ Analysis Complete. Cleaned data saved to 'banggood_cleaned_data.csv'")
```
